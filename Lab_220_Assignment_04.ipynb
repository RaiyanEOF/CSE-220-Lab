## **TASK 01**

```python
class Node_pair:
  def __init__(self, key, value, next = None):
    self.key, self.value, self.next = key, value, next


class Hashtable:

  def __init__(self, size):
    self.ht = [None]*size

  def insert(self, s):
    hashed_index = self.__hash_function(s[0])
    pair = Node_pair(s[0], s[1])
    if self.ht[hashed_index] == None:
      self.ht[hashed_index] = pair
    else:
      pair.next = self.ht[hashed_index]
      self.ht[hashed_index] = pair

  def create_from_array(self, arr):
    for i in arr:
      self.insert(i)

  def print_hashtable(self):
    idx = 0
    for i in self.ht:
      print(idx, ':', end = ' ')
      head = i
      while head != None:
        print(f'({head.key}, {head.value})', end = '-->')
        head = head.next
      print('None')
      idx += 1

  def __hash_function(self, key):
    size = len(self.ht)
    return (key + 3) % size


  def remove(self, key):
    index = self.__hash_function(key)

    head = self.ht[index]

    if head is None:
        return

    if head.key == key:
        self.ht[index] = head.next
        return

    prev = head
    curr = head.next
    while curr is not None:
        if curr.key == key:
            prev.next = curr.next
            return
        prev = curr
        curr = curr.next
    return
```

## **TASK 02**

```python
def evalMathExpression(expr):
    st = Stack()
    for ch in expr:
        if ch in "([{":
            st.push(ch)
        elif ch in ")]}":
            top = st.pop()
            if top is None or (top == '(' and ch != ')') or (top == '[' and ch != ']') or (top == '{' and ch != '}'):
                print("Invalid Expression")
                return
    if not st.isEmpty():
        print("Invalid Expression")
        return

    q = LinkedListQueue()
    prec = {'+':1, '-':1, '*':2, '/':2, '^':3}
    right_assoc = set(['^'])
    i = 0
    n = len(expr)
    while i < n:
        ch = expr[i]
        if ch.isspace():
            i += 1
            continue
        if ch.isdigit():
            num = ch
            i += 1
            while i < n and expr[i].isdigit():
                num += expr[i]
                i += 1
            q.enqueue(num)
            continue
        if ch in "+-*/^":
            while True:
                top = st.peek()
                if top is None:
                    break
                if top in prec:
                    if (ch not in right_assoc and prec[ch] <= prec[top]) or (ch in right_assoc and prec[ch] < prec[top]):
                        q.enqueue(st.pop())
                        continue
                break
            st.push(ch)
            i += 1
            continue
        if ch in "([{":
            st.push(ch)
            i += 1
            continue
        if ch in ")]}":
            need = '(' if ch == ')' else '[' if ch == ']' else '{'
            found = False
            while not st.isEmpty():
                top = st.pop()
                if top == need:
                    found = True
                    break
                q.enqueue(top)
            if not found:
                print("Invalid Expression")
                return
            i += 1
            continue
        i += 1

    while not st.isEmpty():
        top = st.pop()
        if top in "()[]{}":
            print("Invalid Expression")
            return
        q.enqueue(top)

    out = []
    cur = q.front
    while cur is not None:
        out.append(str(cur.elem))
        cur = cur.next
    if len(out) == 0:
        print("Invalid Expression")
        return
    print("Postfix:", " ".join(out))

    while not st.isEmpty():
        st.pop()
    while not q.is_empty():
        tok = q.dequeue()
        if tok.isdigit():
            st.push(int(tok))
        else:
            b = st.pop()
            a = st.pop()
            if a is None or b is None:
                print("Invalid Expression")
                return
            if tok == '+':
                st.push(a + b)
            elif tok == '-':
                st.push(a - b)
            elif tok == '*':
                st.push(a * b)
            elif tok == '/':
                if b == 0:
                    print("Invalid Expression")
                    return
                st.push(int(a / b))
            elif tok == '^':
                st.push(a ** b)
            else:
                print("Invalid Expression")
                return

    res = st.pop()
    if res is None or not st.isEmpty():
        print("Invalid Expression")
        return
    print("Result:", res)
```

## **TASK 03**

```python
def removeConsecutiveDups(s):
    q = LinkedListQueue()
    for ch in s:
        q.enqueue(ch)
    if q.is_empty():
        return ""

    prev = q.dequeue()
    result = prev

    while not q.is_empty():
        curr = q.dequeue()
        if curr != prev:
            result += curr
        prev = curr

    return result
```
