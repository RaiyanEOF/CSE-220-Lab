### **Node Class**

```python
class Node:
  def __init__(self,elem,next = None):
    self.elem,self.next = elem,next

def createList(arr):
  head = Node(arr[0])
  tail = head
  for i in range(1,len(arr)):
    newNode = Node(arr[i])
    tail.next = newNode
    tail = newNode
  return head

def printLinkedList(head):
  temp = head
  while temp != None:
    if temp.next != None:
      print(temp.elem, end = '-->')
    else:
      print(temp.elem)
    temp = temp.next
  print()
```

# **TASK 02**

```python
import numpy as np


class Node:
    def __init__(self, data):
        self.data = data
        self.next = None


def createList(arr):
    if len(arr) == 0:
        return None

    head = Node(arr[0])
    temp = head
    for i in range(1, len(arr)):
        newNode = Node(arr[i])
        temp.next = newNode
        temp = newNode
    return head


def reverse_linked_list(head):
    prev = None
    curr = head
    while curr is not None:
        next_node = curr.next
        curr.next = prev
        prev = curr
        curr = next_node
    return prev


def printLinkedList(head):
    temp = head
    while temp is not None:
        if temp.next is None:
            print(temp.data, end="")
        else:
            print(temp.data, end="→")
        temp = temp.next
    print()


def word_Decoder(head):
    temp = head
    index = 0
    decoded_head = None
    decoded_tail = None

    while temp is not None:
        if index != 0 and index % 3 == 0:
            new_node = Node(temp.data)
            if decoded_head is None:
                decoded_head = new_node
                decoded_tail = new_node
            else:
                decoded_tail.next = new_node
                decoded_tail = new_node
        index += 1
        temp = temp.next


    decoded_head = reverse_linked_list(decoded_head)

    dummy = Node("None")
    dummy.next = decoded_head
    return dummy


# [DO NOT MODIFY THE TESTER CODES BELOW]
# [THERE WILL BE 50% PENALTY IF IT'S MODIFIED]


# ================== TEST CASES ==================
print('==============Test Case 1=============')
head = createList(np.array(['B', 'M', 'D', 'T', 'N', 'O', 'A', 'P', 'S', 'C']))
print("Encoded Word:")
printLinkedList(head)   # B→M→D→T→N→O→A→P→S→C

result = word_Decoder(head)
print("Decoded Word:")
printLinkedList(result) # Expected: C→A→T


print('==============Test Case 2=============')
head = createList(np.array(['Z', 'O', 'T', 'N', 'X']))
print("Encoded Word:")
printLinkedList(head)   # Z→O→T→N→X

result = word_Decoder(head)
print("Decoded Word:")
printLinkedList(result) # Expected: N
```
