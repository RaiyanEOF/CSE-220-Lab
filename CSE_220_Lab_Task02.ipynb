### **Node Class**

```python
class Node:
  def __init__(self,elem,next = None):
    self.elem,self.next = elem,next

def createList(arr):
  head = Node(arr[0])
  tail = head
  for i in range(1,len(arr)):
    newNode = Node(arr[i])
    tail.next = newNode
    tail = newNode
  return head

def printLinkedList(head):
  temp = head
  while temp != None:
    if temp.next != None:
      print(temp.elem, end = '-->')
    else:
      print(temp.elem)
    temp = temp.next
  print()
```

# **TASK 02**

```python
import numpy as np


class Node:
    def __init__(self, data):
        self.data = data
        self.next = None


def createList(arr):
    if len(arr) == 0:
        return None

    head = Node(arr[0])
    temp = head
    for i in range(1, len(arr)):
        newNode = Node(arr[i])
        temp.next = newNode
        temp = newNode
    return head



def printLinkedList(head):
    temp = head
    while temp is not None:
        if temp.next is None:
            print(temp.data, end="")
        else:
            print(temp.data, end="→")
        temp = temp.next
    print()


def word_Decoder(head):
    # Step 1: Find length of linked list
    temp = head
    length = 0
    while temp is not None:
        length += 1
        temp = temp.next

    temp = head
    index = 0
    picked = []
    while temp is not None:
        if index != 0 and index % 3 == 0:
            picked.append(temp.data)
        index += 1
        temp = temp.next
    reversed_picked = []
    for i in range(len(picked)-1, -1, -1):
        reversed_picked.append(picked[i])

    dummy = Node("None")
    temp = dummy
    for ch in reversed_picked:
        newNode = Node(ch)
        temp.next = newNode
        temp = newNode

    return dummy



print('==============Test Case 1=============')
head = createList(np.array(['B', 'M', 'D', 'T', 'N', 'O', 'A', 'P', 'S', 'C']))
print("Encoded Word:")
printLinkedList(head)   # B→M→D→T→N→O→A→P→S→C→None

result = word_Decoder(head)
print("Decoded Word:")
printLinkedList(result) # None→C→A→T→None


print('==============Test Case 2=============')
head = createList(np.array(['Z', 'O', 'T', 'N', 'X']))
print("Encoded Word:")
printLinkedList(head)   # Z→O→T→N→X→None

result = word_Decoder(head)
print("Decoded Word:")
printLinkedList(result) # None→N→None

```
