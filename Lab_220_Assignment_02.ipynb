## **Node Class**

```python
import numpy as np
class Node:
  def __init__(self,elem,next = None):
    self.elem = elem
    self.next = next

def createList(arr):
  head = Node(arr[0])
  tail = head
  for i in range(1,len(arr)):
    newNode = Node(arr[i])
    tail.next = newNode
    tail = newNode
  return head

def printLinkedList(head):
  temp = head
  while temp != None:
    if temp.next != None:
      print(temp.elem, end = '-->')
    else:
      print(temp.elem)
    temp = temp.next
  print()
```

## **Task 01**

```python
import numpy as np
def check_similar(building_1, building_2):
 if building_1.elem == building_2.elem:
  return True
 else:
  return False


building_1 = createList(np.array(['Red', 'Green', 'Yellow', 'Red', 'Blue', 'Green']))
building_2 = createList(np.array(['Red', 'Green', 'Yellow', 'Red', 'Blue', 'Green']))

returned_value = check_similar(building_1, building_2)
print(returned_value)

if(returned_value):
  print('Similar')
else:
  print('Not Similar')
```

## **TASK 02**

```python
def organizeBooks(head, arr):
    if head is None:
        return None

    current = head
    index1 = 0

    while current:
        max_node = current
        max_index = index1

        temp = current.next
        index2 = index1 + 1

        while temp:
            if arr[index2] > arr[max_index]:
                max_node = temp
                max_index = index2
            temp = temp.next
            index2 += 1

        if max_node is not current:
            arr[index1], arr[max_index] = arr[max_index], arr[index1]
            current.elem, max_node.elem = max_node.elem, current.elem

        current = current.next
        index1 += 1

    return head



books1 = createList(np.array(["Dune", "IT", "Coraline", "Inferno", "Twilight"]))
popularity1 = np.array([8, 10, 5, 10, 6])
books = organizeBooks(books1,popularity1)
printLinkedList(books)
print(popularity1)
```

## **TASK 03**

```python
def alternate_merge(head1, head2):
  if head1 is None :
    return head2

  if head2 is None :
    return head1

  curr1 = head1
  curr2 = head2

  while curr1 and curr2:
      next1 = curr1.next
      next2 = curr2.next

      curr1.next = curr2
      curr2.next = next1

      curr1 = next1
      curr2 = next2

  return head1



head1 = createList(np.array([1,2,6,8,11]))
head2 = createList(np.array([5,7,3,9,4]))
head = alternate_merge(head1,head2)
printLinkedList(head)
```

## **TASK 04**

```python
import numpy as np

def idGenerator(head1, head2, head3):

    if head1 is None:
        return None

    head_out = None
    curr1 = head1
    while curr1 is not None:

        new_node = Node(curr1.elem)

        new_node.next = head_out
        head_out = new_node

        curr1 = curr1.next


    tail = head_out

    if tail is None:
        tail = None

    else:
        while tail.next is not None:
            tail = tail.next


    c2 = head2
    c3 = head3


    while c2 is not None and c3 is not None:
        s = c2.elem + c3.elem
        if s >= 10:
            s = s % 10

        new_digit = Node(s)

        if head_out is None:
            head_out = new_digit
            tail = new_digit
        else:
            tail.next = new_digit
            tail = new_digit

        c2 = c2.next
        c3 = c3.next

    if tail is not None:
        tail.next = None

    return head_out

head1 = createList(np.array([0,3,2,2]))
head2 = createList(np.array([5,2,2,1]))
head3 = createList(np.array([4,3,2,1]))

result = idGenerator(head1, head2, head3)

print("Linked List 1:")
printLinkedList(head1)
print("Linked List 2:")
printLinkedList(head2)
print("Linked List 3:")
printLinkedList(head3)

print("New ID:")
printLinkedList(result)
```
